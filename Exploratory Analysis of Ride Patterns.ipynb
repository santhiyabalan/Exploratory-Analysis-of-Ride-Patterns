```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('/content/uber_rides_tamilnadu.csv')
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
# Step 4: Check dimensions of the dataset
print(f"Dataset Shape: {df.shape}")  # (Rows, Columns)
```

```python
#Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
 #Drop rows with missing values
df_cleaned= df.dropna()
```

```python
#remove duplicates
df_cleaned=df_cleaned.drop_duplicates()
```

```python
#Convert data types (for example, converting 'pickup_time' to datetime)
df_cleaned['pickup_time'] = pd.to_datetime(df_cleaned['pickup_time'], dayfirst=True)
```

```python
df_cleaned['dropoff_time'] = pd.to_datetime(df_cleaned['dropoff_time'],dayfirst=True)
```

```python

print(df_cleaned['pickup_time'].head())
```

```python
#Check if data types are now correct
df_cleaned.info()
```

```python
#create new columns (e.g., extract day from pickup_time)
df_cleaned['day_of_week'] = df_cleaned['pickup_time'].dt.day_name()

df_cleaned.head()
```

## Exploratory Data Analysis (EDA)

```python
# Step 1: Summary statistics
df_cleaned.describe()  # Get statistical summary for numerical columns
```

```python
# Categorize rides based on duration (in minutes)
df_cleaned['ride_category'] = pd.cut(
    df_cleaned['ride_duration'],
    bins=[0, 10, 30, float('inf')],  # Define bins: short (0-10 mins), medium (10-30 mins), long (>30 mins)
    labels=['short', 'medium', 'long']
)

 #Preview the data
df_cleaned[['ride_duration','ride_category']].head()
```

```python
import matplotlib.pyplot as plt
# Filter only numerical columns
numerical_cols=df_cleaned.select_dtypes(include=['float64','int64'])
# Compute the correlation matrix
corr_matrix = numerical_cols.corr()


```

# Data Visualization

```python

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(5, 3))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
 #Distribution of Payment Methods
plt.figure(figsize=(5, 3))
sns.countplot(x='payment_method', data=df_cleaned)
plt.title("Distribution of Payment Methods")
plt.xlabel("Payment Method")
plt.ylabel("Ride Count")
plt.show()
```

```python
# Step 3: Fare vs. Ride Duration
plt.figure(figsize=(5, 3))
sns.scatterplot(x='ride_duration', y='fare_amount', data=df_cleaned)
plt.title("Fare vs. Ride Duration")
plt.xlabel("Ride Duration (minutes)")
plt.ylabel("Fare Amount")
plt.show()
```

```python
#Fare Distribution Analysis by Passenger Count and Ride Duration"
plt.figure(figsize=(10, 6))
sns.boxplot(x='passenger_count', y='fare_amount', data=df_cleaned, hue='ride_duration')
plt.title("Fare Amount by Passenger Count and Ride Duration")
plt.xlabel("Passenger Count")
plt.ylabel("Fare Amount")
plt.show()
```

```python
# Step 5: Heatmap of Pickup Locations
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pickup_longitude', y='pickup_latitude', data=df_cleaned, hue='payment_method', palette="viridis")
plt.title("Heatmap of Pickup Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Payment Method")
plt.show()
```

## Answering Business Questions and Generating Insights

```python
#counting rides per day and sorting by frequency

df_cleaned['pickup_date'] = df_cleaned['pickup_time'].dt.date
ride_count_per_day = df_cleaned['pickup_date'].value_counts().sort_values(ascending=False)
print(ride_count_per_day)

```

```python
#Average Ride Duration by Pickup Location
# Calculate the average ride duration by pickup location
avg_duration_by_location = df_cleaned.groupby('pickup_location')['ride_duration'].mean()
print(avg_duration_by_location)
```

```python
 #Get the count of rides by day of the week
day_counts = df_cleaned['day_of_week'].value_counts().sort_index()
print(day_counts)
```

```python
# Plot the frequency by day of the week
plt.figure(figsize=(5, 3))
day_counts.plot(kind='bar')
plt.title("Ride Frequency by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Ride Count")
plt.show()
```

```python
 #Regions with the Most Rides
# Get the top 10 pickup locations with the most rides
top_regions = df_cleaned['pickup_location'].value_counts().head(10)
print(top_regions)
```

```python
# Plot the most frequent pickup locations
plt.figure(figsize=(5, 3))
top_regions.plot(kind='bar')
plt.title("Top Pickup Locations")
plt.xlabel("Pickup Location")
plt.ylabel("Ride Count")
plt.show()
```

```python
 #Step 5: Average Fare by Time of Day (skip if no hour information)
# Since no hour information, you can group by date for average fare analysis
avg_fare_by_date = df_cleaned.groupby('pickup_date')['fare_amount'].mean()
print(avg_fare_by_date)
```

```python
# Plot the average fare by date
plt.figure(figsize=(10, 6))
avg_fare_by_date.plot(kind='line')
plt.title("Average Fare by Date")
plt.xlabel("Date")
plt.ylabel("Average Fare")
plt.show()
```

```python
avg_fare_by_location = df_cleaned.groupby('pickup_location')['fare_amount'].mean().sort_values(ascending=False)
```

```python
 #Print the top locations with average fare
print(avg_fare_by_location.head(10))
```

```python
avg_fare_by_payment = df_cleaned.groupby('payment_method')['fare_amount'].mean()
print(avg_fare_by_payment)
```

```python
# Step 4: Rides frequency by pickup location
ride_frequency_by_location = df_cleaned['pickup_location'].value_counts().head(10)
# Step 5: Average fare by pickup location
avg_fare_by_location = df_cleaned.groupby('pickup_location')['fare_amount'].mean().sort_values(ascending=False).head(10)

# Plot the ride frequency and average fare for top 10 locations
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.set_xlabel('Pickup Location')
ax1.set_ylabel('Ride Frequency', color='tab:blue')
ride_frequency_by_location.plot(kind='bar', ax=ax1, color='tab:blue', position=1, width=0.4)

ax2 = ax1.twinx()
ax2.set_ylabel('Average Fare Amount', color='tab:orange')
avg_fare_by_location.plot(kind='bar', ax=ax2, color='tab:orange', position=0, width=0.4)

plt.title("Ride Frequency and Average Fare by Pickup Location")
plt.xticks(rotation=45, ha='right')
plt.show()
```
